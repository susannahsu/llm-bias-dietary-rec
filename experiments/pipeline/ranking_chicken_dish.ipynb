```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import re, pickle, os, csv, ast

from openai import OpenAI
from dotenv import load_dotenv
```

```python
# Load API key from .env file
load_dotenv()

key = os.getenv('OPENAI_API_KEY')
if key is None:
    raise ValueError("The OPENAI_API_KEY environment variable is not set \
                     or .env file is missing.")

client = OpenAI(
    api_key=key
)
```

```python
def call_openai_api(user_prompt, n_runs=1, model="gpt-4-turbo-2024-04-09"):
    responses = []
    for run_number in range(1, n_runs + 1):
        completion = client.chat.completions.create(
            model=model,
            messages=[
                # {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt},
            ],
        )
        response_content = completion.choices[0].message.content
        print(response_content)
        print("========================================next call")
        responses.append(response_content)
    return responses
```

```python
def load_prompt_from_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.read()

def generate_user_prompt(file_path):
    return load_prompt_from_file(file_path)

def generate_system_prompt(file_path):
    return load_prompt_from_file(file_path)
```

```python
user_prompt_path = './prompts/rank_chicken_dish/rank_chicken_dish_user_prompt.txt'
user_prompt = generate_user_prompt(user_prompt_path)
```

```python
call_openai_api(user_prompt)
```

```python
chicken_rank_resp = call_openai_api(user_prompt, 30)
```

```python
with open('./responses/rank_chicken_dish/chicken_rank_resp.pkl', 'wb') as f:
    pickle.dump(chicken_rank_resp, f)
```

```python
with open('./responses/rank_chicken_dish/chicken_rank_resp.pkl', 'rb') as f:
    chicken_rank_resp = pickle.load(f)
```
